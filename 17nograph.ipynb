```python
column_names=['Age','Sex','cp','trestbp','chol','fbs','restecg','thalah','exang','outpeak','slope0','ca','thal','target']
```

```python
file_path='Switzerland.csv'
```

```python
df=pd.read_csv('Switzerland.csv',names=column_names , na_values="?")
print(df)
```

```python
df.isnull().sum()
```

```python
df['fbs'].fillna(0,inplace=True)
print(df.isnull().sum())
```

```python
#Histogram

import matplotlib.pyplot as plt
```

```python
df.hist(figsize=(10,15))
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(10,15))
plt.hist(df['Age'],bins=10,color='blue',edgecolor='red')
plt.title("Graph")
plt.xlabel("Age")
plt.ylabel('Frequence')
plt.grid(True)
plt.show()
```

```python
df.plot(kind="scatter",x="Age",y="cp",figsize=(5,8))
plt.title('Age vs. Max Heart Rate')
plt.xlabel('Age')
plt.ylabel('cp')
plt.grid(True)
plt.show()
```

```python
chest=df['cp'].value_counts()
chest.plot(kind="bar",figsize=(5,7))
plt.title('cp vs count')
plt.xlabel('cp')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()
```

```python
chol=df.groupby('Age')['chol'].mean()
chol.plot(kind='line',figsize=(10,6),color='yellow')
plt.title('Average Serum Cholesterol by Age')
plt.xlabel('Age')
plt.ylabel('Average Serum Cholesterol (mg/dl)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
df['Age'].plot(kind='box')
plt.title('Age Distribution')

plt.subplot(2, 2, 2)
df.plot(kind='scatter', x='Age', y='chol')
plt.title('Age vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol (mg/dl)')

plt.subplot(2, 2, 3)
df['cp'].plot(kind='box')
plt.title('Max Heart Rate Distribution')

plt.subplot(2, 2, 4)
df.plot(kind='scatter', x='cp', y='chol')
plt.title('Max Heart Rate vs. Cholesterol')
plt.xlabel('Max Heart Rate')
plt.ylabel('Cholesterol (mg/dl)')

plt.tight_layout()
plt.show()
```
